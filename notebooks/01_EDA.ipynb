```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
df = pd.read_csv('../data/PS2_Dataset.csv')
print(df.shape)
print(df.head())
```

```python
print("Column names:\n", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(12, 5))
df['Suggested Job Role'].value_counts().plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Distribution of Career Roles', fontsize=14)
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
num_cols = ['Logical quotient rating', 'hackathons','coding skills rating', 'public speaking points']

df[num_cols].hist(figsize=(10, 6), bins=10, color='orange', edgecolor='black')
plt.suptitle('Numeric Features Distribution', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns.drop('Suggested Job Role')

fig, axes = plt.subplots(4, 4, figsize=(18, 14))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    df[col].value_counts().plot(kind='bar', ax=axes[i], color='mediumseagreen', edgecolor='black')
    axes[i].set_title(col, fontsize=9)
    axes[i].tick_params(axis='x', rotation=30)

plt.suptitle('Categorical Features Distribution', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Numeric Features')
plt.tight_layout()
plt.show()
```
